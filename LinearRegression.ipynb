```python
# y = m*x + b
# where m is the slope and b is the y-intercept

# E = 1/n * Σ (yi - (m*xi + b))^2
# where E is the error, n is the number of data points, and (xi, yi) are the data points
# Also known as Mean Squared Error (MSE)
# We want to minimize E by adjusting m and b


# Gradient Descent
# m = m - α * dE/dm
# b = b - α * dE/db
# where α is the learning rate, dE/dm is the derivative of E with respect to m, and dE/db is the derivative of E with respect to b
# dE/dm = -2/n * Σ xi * (yi - (m*xi + b))
# dE/db = -2/n * Σ (yi - (m*xi + b))

# m = m - α * dE/dm
# b = b - α * dE/db
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv('datasets/LinearRegression.csv')
dataX = data['x']

dataY = data['y']


plt.scatter(dataX, dataY)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter plot of X vs Y')
plt.show()
```

```python
def loss_function(m,b,X,Y):
    total_error = 0

    for i in range(len(X)):
        x = X[i]
        y = Y[i]

        predicted_value = m*x + b

        total_error = (y - predicted_value)**2

    
    total_error = total_error/float(len(X))

    return total_error



def gradient_descent(m_now, b_now, X, Y, alpha):

    m_gradient = 0
    b_gradient = 0

    n = len(X)

    for i in range(n):
        x = X[i]
        y = Y[i]

        predicted_value = m_now*x + b_now

        m_gradient += x * (y - predicted_value)

        b_gradient += (y - predicted_value)
        
    
    m_gradient = m_gradient * float(-2/n)
    b_gradient = b_gradient * float(-2/n)


    m = m_now - alpha*m_gradient
    b = b_now - alpha*b_gradient

    return m,b



        

        




```

```python
m = 0
b = 0
alpha = 0.001
epochs = 1000

for i in range(epochs):
    m, b = gradient_descent(m,b, dataX, dataY, alpha)



print(m,b)
plt.scatter(dataX,dataY, color = "red")


plt.plot(list(range(0,10)), [m*x + b for x in range(0,10)])
```
